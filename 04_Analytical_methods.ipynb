The content and notation follows the corresponding tutorial of the course. An isotropic stiffness is represented as $\mathbb{C} = \mathrm{iso}(c_1,c_2) = c_1 \mathbb{P}_1 + c_2 \mathbb{P}_2$ and has the eigenvalues $c_1 = 3 K$ and $c_2 = 2 G$, where $K$ and $G$ correspond to the compression and shear modulus of the material.
